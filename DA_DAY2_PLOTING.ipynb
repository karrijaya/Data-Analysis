```python
#runs scored by 10 new players [100,50,91,78,89,25,34,19,9,10] wickets taken by same 10 new players[1,0,2,0,3,7,8,9,7,5]
#form clusters for batsmen and bolwersruns=np.array([100,50,91,78,89,25,34,19,9,10])
wickets=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(runs,wickets,color="m",marker="p")
plt.show()
```

```python
#plot the scores of two studentsin five subjects,subjects as x-axis and marks as y-axis
s1=np.array([89,92,73,85,95])
s2=np.array([99,97,82,79,91])
sub=np.array(["maths","science","social","telugu","english"])
plt.plot(sub,s1,label="s1")
plt.plot(sub,s2,label="s2")
plt.title("marks v/s subjects")
plt.legend()
plt.show()
```

```python
stu1=[56,78,92,95,78]
stu2=[77,89,91,23,45]
sub=["maths","science","english","telugu","social"]
plt.subplot(2,1,1)
plt.plot(sub,stu1,label="student1",color="m")
plt.legend()
plt.subplot(2,1,2)
plt.plot(sub,stu2,label="student2",color="c")
plt.legend()
plt.show()
```

```python
A=np.array([230,560,780,127,128])
B=np.array([200,160,270,127,400])
print("Revenue of companyA ",A)
print("Revenue of companyB ",B)
x=np.diff(A)
print("profits of A",x)
y=np.diff(B)
print("profits of B",y)
year=np.array(["2019-2020","2020-2021","2021-2022","2022-2023"])
plt.subplot(1,2,1)
plt.bar(year,x,label="companyA",color="k")
plt.legend(loc="best")
plt.title("profits of company A")

plt.subplot(1,2,2)
plt.bar(year,y,label="companyB",color="green")
plt.legend(loc="best")
plt.title("profits of company B")
```

```python
a=np.array([25,60,5,10])
labe=["AIML","Python","Pandas","Numpy"]
plt.pie(a,labels=labe)
plt.show()
```

```python
a=np.array([25,60,5,10])
labe=["AIML","Python","Pandas","Numpy"]
explo=[0.1,0.1,0.1,0.1]
plt.pie(a,labels=labe,explode=explo,startangle=180,shadow=True)
plt.legend(loc="best")
plt.show()
```
